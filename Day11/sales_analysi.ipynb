"""
Analyse a sales CSV: find total revenue per product and best-selling month
"""

```python


import pandas as pd
from datetime import datetime 
```

```python
df = pd.read_csv('./csv/100 Sales Records.csv')
print(df.head())
```

```python
# total_revenue_by_item = df.groupby('Item Type')['Total Revenue'].sum()

total_revenue_by_item = df.groupby('Item Type').agg(
    Total_revenue = ('Total Revenue', 'sum')
).reset_index()

print("Total Revenue By Item : ")
print(total_revenue_by_item)
```

```python
# Converting order date into 'datetime' , and 
# making new column month

df['Order Date'] = pd.to_datetime(df['Order Date'])
df['month'] = df['Order Date'].dt.month
```

```python
# Finding Month With Highest total  Revenue

revenues = df.groupby('month')['Total Revenue'].sum()

highest_revenue = revenues.max()
best_selling_month = revenues.idxmax()

print("Best Selling Month with Highest Revenue is : ",best_selling_month)
print("It's  Highest Revenue is : ",highest_revenue)
```
