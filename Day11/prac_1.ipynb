#### Read and anaysis of CSV 

```python
df = pd.read_csv('./csv/movie2.csv')
```

```python
df.head()
```

```python
# Max revenue movie name
df.loc[df['Revenue (Millions)'].idxmax()]
```

```python
# first 10 rows
df[:10]
```

```python
df.info()
```

```python
# FIle all none with  0
df.fillna(0)
```

```python
# count none value in colums
df.isnull().sum() 

# movies after 2015 release
df[df['Year'] > 2015]

# movies with >8 rating and >200 million box office
df[(df['Rating'] > 8) & (df['Revenue (Millions)'] > 200)]
```

```python
# Genre that contains 'Action'
df[df['Genre'].str.contains('Action')]
```

```python
# Top 5 Highest grossing movies
df.sort_values(by='Revenue (Millions)', ascending=False)[:5]
```

```python
# Highest Rated top 10 movies
df.sort_values(by='Rating', ascending=False)[:10]
```

```python
# Lowest runtime movie
df.sort_values(by='Runtime (Minutes)').iloc[0]
```

```python
# Avg. rating year wise
df.groupby('Year')['Rating'].mean()
```

```python
# total revenue of movies year wise
df.groupby('Year')['Revenue (Millions)'].sum()
```

```python
# Find the director with the highest average rating.
df.groupby('Director')['Rating'].mean().sort_values(ascending=False).head(1)
```

```python
# # Which director has made the most movies?

# df.groupby('Director').count()['Title'].sort_values(ascending=False).head(1)

df['Director'].value_counts()
```

```python
# Find the most common genre.
df['Genre'].str.split(',').explode().value_counts()
```

```python
# Find movies with votes > 500000.
df.query('Votes > 500000')[:10]
# df[df['Votes'] > 500000]
```

```python
df.head()
```

```python
# Task :-
# Load a CSV dataset, calculate mean/max/min per group, generate a simple bar chart

df.groupby('Year').agg(
    avg_revenue=('Revenue (Millions)', 'mean'),
    min_revenue=('Revenue (Millions)', 'min'),
    max_revenue=('Revenue (Millions)', 'max')
)
```

## Bar Chart with pandas

```python
avg_rating = df.groupby('Year')['Revenue (Millions)'].max()
avg_rating.plot(kind='bar')
```

```python
avg_rating = df.groupby('Year')['Revenue (Millions)'].mean()
avg_rating.plot(kind='bar')
```
